**1. How image thresholding helps in each application mentioned in the materials?**  

Ans: 1) Segmentation –  Thresholding separates objects from the background, making it easier to analyze specific regions. It is widely used in medical imaging, such as detecting tumors in X-rays and MRIs.


2) Object Detection – Helps isolate objects of interest by converting images into high-contrast binary forms. This is useful in applications like traffic monitoring, handwriting recognition, and industrial automation.


3) Edge Detection –  Enhances object boundaries by highlighting differences in pixel intensity. It is often used in structural inspections, crack detection, and feature extraction.


 4) Image Enhancement – Improves contrast and clarity by adjusting pixel values, making images more readable. This is beneficial in medical scans, satellite imaging, and low-light image processing.

5) Pattern Recognition – Simplifies image data, enabling better classification of objects. It is used in applications like optical character recognition (OCR), barcode scanning, and facial recognition systems.

**2. Why the grayscale image showed color when plotted without using cmap=’gray’? Explain.**

Ans:   When a grayscale image is plotted without specifying cmap='gray', it may appear in color due to the way Matplotlib (or other plotting libraries) handle image visualization. Here's why:

1. Default Colormap

When using plt.imshow(image), Matplotlib automatically applies a default colormap (viridis or another default one) to the image.
A grayscale image is essentially a single-channel image, meaning each pixel has only one intensity value.
However, imshow() assumes multi-channel (RGB) data unless instructed otherwise. It maps the intensity values to a default colormap, creating a false-color representation.

2. Interpreting 2D vs. 3D Data

If the image is 2D (shape: (height, width)), imshow() treats it as a single-channel image but applies a colormap to visualize intensity variations.
If the image is 3D (shape: (height, width, 3) or (height, width, 4)), it is interpreted as an RGB(A) image and displayed as true color.

Solution :
To display a grayscale image correctly, explicitly set the colormap:

```python
image = cv2.imread('/content/drive/MyDrive/Innovative_Skills/session_3/pexels-ivan-samkov-4240503.jpg', cv2.IMREAD_GRAYSCALE)


plt.subplot(1, 2, 1)
plt.imshow(image)
plt.title("Without cmap='gray'")


plt.subplot(1, 2, 2)
plt.imshow(image, cmap='gray')
plt.title("With cmap='gray'")

plt.show()
```

**1. Take three images of your choice and apply different global thresholding on them.**

```python
img = cv2.imread('/content/drive/MyDrive/Innovative_Skills/session_3/pexels-fauxels-3184460.jpg', cv2.IMREAD_GRAYSCALE)

ret, thresholded = cv2.threshold(img, 150, 255,  cv2.THRESH_TOZERO_INV)


plt.subplot(1, 2, 1)
plt.imshow(img, cmap='gray')
plt.title("Without Threshold")


plt.subplot(1, 2, 2)
plt.imshow(thresholded, cmap='gray')
plt.title("With Threshold")
print(ret)
```
