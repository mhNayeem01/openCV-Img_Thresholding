## Adaptive Thresholding

```python
img.shape
```

```python
adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 5, 10)
plt.imshow(adaptive_mean, 'gray')
```

```python
adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 15, 2)
plt.imshow(adaptive_mean, 'gray')
```

```python
adaptive_mean = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 25, 2)
plt.imshow(adaptive_mean, 'gray')
```

```python
adaptive_gaussian = cv.adaptiveThreshold(img, 255, cv.ADAPTIVE_THRESH_GAUSSIAN_C, cv.THRESH_BINARY, 11, 2)
plt.imshow(adaptive_gaussian, 'gray')
```

## Otsu's Thresholding

```python
ret1, otsu1 = cv2.threshold(img, 160, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)
ret2, otsu2 = cv2.threshold(img, 60, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

plt.subplot(1, 2, 1)
plt.imshow(otsu1, cmap='gray')
# plt.title('Otsu Thresholding (v=160)')

plt.subplot(1, 2, 2)
plt.imshow(otsu2, cmap='gray')
# plt.title('Otsu Thresholding (v=60)')

plt.show()
```
